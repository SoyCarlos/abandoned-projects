```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

from plotly.offline import init_notebook_mode, iplot
from IPython.display import display, HTML
import plotly.graph_objs as go


init_notebook_mode(connected=True)
```

```python
def split_year(inp):
    split = inp.split("(")
    if len(split) == 1:
        return np.nan
    return split[1].split(')')[0]
```

```python
_2013 = pd.read_csv("data/2012-2013 ABSA.csv")
_2013['Allocation'] = _2013['Allocation'].apply(lambda x: x.replace(",","").strip("$"))
_2013.rename(columns={'Years':'Years of Sponsorship'}, inplace=True)
_2013['Allocation'] = pd.to_numeric(_2013['Allocation'],errors='coerce')
_2013.head()
```

```python
_2014 = pd.read_csv("data/2013-2014 ABSA.csv")
_2014['Allocation'] = _2014['Allocation'].apply(lambda x: x.replace(",","").strip("$"))
_2014.rename(columns={'Years':'Years of Sponsorship'}, inplace=True)
_2014['Allocation'] = pd.to_numeric(_2014['Allocation'],errors='coerce')
_2014.head()
```

```python
_2015 = pd.read_csv("data/2014-2015 ABSA.csv")
_2015['Final Allocation'] = _2015['Final Allocation'].apply(lambda x: x.replace(",","").strip("$"))
_2015.rename(columns={'Final Allocation': 'Allocation'}, inplace=True)
_2015['Allocation'] = pd.to_numeric(_2015['Allocation'],errors='coerce')
_2015.head()
```

```python
_2016 = pd.read_csv("data/2015-2016 ABSA.csv")
_2016['FINAL ALLOCATION'] = _2016['FINAL ALLOCATION'].apply(lambda x: x.replace(",","").strip("$"))
_2016.rename(columns={'FINAL ALLOCATION': 'Allocation'}, inplace=True)
_2016['Type'] = _2016['TYPE (YEAR)'].apply(lambda x: x.split("(")[0])
_2016['Years of Sponsorship'] = _2016['TYPE (YEAR)'].apply(lambda x: split_year(x))
_2016.drop('TYPE (YEAR)', axis=1, inplace=True)
_2016['Allocation'] = pd.to_numeric(_2016['Allocation'],errors='coerce')
_2016.head()
```

```python
_2017 = pd.read_csv("data/2016-2017 ABSA.csv")
_2017['Allocation'] = _2017['Allocation'].apply(lambda x: x.replace(",","").strip("$"))
_2017['Type'] = _2017['TYPE (YEAR)'].apply(lambda x: x.split("(")[0])
_2017['Years of Sponsorship'] = _2017['TYPE (YEAR)'].apply(lambda x: split_year(x))
_2017.drop('TYPE (YEAR)', axis=1, inplace=True)
_2017['Allocation'] = pd.to_numeric(_2017['Allocation'],errors='coerce')
_2017.head()
```

```python
_2018 = pd.read_csv("data/2017-2018 ABSA.csv")
_2018['Initial Allocation'] = _2018['Initial Allocation'].apply(lambda x: x.replace(",","").strip("$"))
_2018.rename(columns={'Year':'Years of Sponsorship', 'Initial Allocation': 'Allocation', 'Sponsorship Category':'Type'}, inplace=True)
_2018['Allocation'] = pd.to_numeric(_2018['Allocation'],errors='coerce')
_2018.head()
```

```python
_2019 = pd.read_csv("data/2018-2019 ABSA.csv")
_2019['Final Allocation'] = _2019['Final Allocation'].apply(lambda x: x.replace(",","").strip("$"))
_2019.rename(columns={'Years':'Years of Sponsorship', 'Final Allocation': 'Allocation', 'Sponsorship Category':'Type'}, inplace=True)
_2019['Allocation'] = pd.to_numeric(_2019['Allocation'],errors='coerce')
_2019.head()
```

```python
_2020 = pd.read_csv("data/2019-2020 ABSA.csv")
_2020['Allocation'] = _2020['Allocation'].apply(lambda x: x.replace(",","").strip("$"))
_2020.rename(columns={'Years':'Years of Sponsorship', 'Final Allocation': 'Allocation', 'Sponsorship Category':'Type'}, inplace=True)
_2020['Allocation'] = pd.to_numeric(_2020['Allocation'],errors='coerce')
_2020.head()
```
