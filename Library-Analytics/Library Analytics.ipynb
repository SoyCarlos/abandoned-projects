```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Helper Funcs

#Get day of the week (Integer)
get_day = lambda date: date.dayofweek

#Convert integer day of week to string
def convert_date(date):
    dates = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
    return dates.get(date)

#Assign whethere it is a week day (Should be a school day)
def week_day(date):
    dates = {0:'Yes', 1:'Yes', 2:'Yes', 3:'Yes', 4:'Yes', 5:'No', 6:'No'}
    return dates.get(date)
```

```python
daily_agg_17_18 = pd.read_csv("daily_aggregate-2017-2018.csv")
original = pd.read_csv("daily_aggregate-2017-2018.csv")
daily_agg_17_18["Day"] = pd.to_datetime(daily_agg_17_18["Day"])
daily_agg_17_18["Day of Week (INT)"] = daily_agg_17_18['Day'].apply(get_day)
daily_agg_17_18["Day of Week"] = daily_agg_17_18['Day of Week (INT)'].apply(convert_date)
daily_agg_17_18["Week Day"] = daily_agg_17_18['Day of Week (INT)'].apply(week_day)
daily_agg_17_18.groupby("Facility").count()
```

```python
bancroft = daily_agg_17_18[daily_agg_17_18["Facility"] == "BANC Library"]
doe = daily_agg_17_18[daily_agg_17_18["Facility"] == "DOE Library"]
stacks = daily_agg_17_18[daily_agg_17_18["Facility"] == "DOE STACKS"]
moffit = daily_agg_17_18[daily_agg_17_18["Facility"] == "MOFF Library"]
```

```python
plt.plot(moffit['Day'], moffit['Count'])
plt.xticks(rotation='vertical')
plt.title("Moffit Library")
plt.show()
```

```python
plt.plot(doe['Day'], doe['Count'])
plt.xticks(rotation='vertical')
plt.title("Doe Library")
plt.show()
```

```python
plt.plot(stacks['Day'], stacks['Count'])
plt.xticks(rotation='vertical')
plt.title("Stacks Library")
plt.show()
```

```python
plt.plot(bancroft['Day'], bancroft['Count'])
plt.xticks(rotation='vertical')
plt.title("Bancroft Library")
plt.show()
```
